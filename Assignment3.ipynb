```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("spam.csv", encoding='latin-1')
```

```python
add=df.head(10)
add
```

```python
data = add.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
data.head(10)
```

```python
data=data.rename(columns={"v1":"Category","v2":"Message"})
data
```

```python
data['Spam']=data['Category'].apply(lambda x:1 if x=='spam' else 0)
```

```python
data
```

```python
%matplotlib inline
data.hist(by='Category',column='Spam',bins=10,figsize=[15,6]) 
```
